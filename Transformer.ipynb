## Sentiment Analysis

```python
#!pip install transformers
```

```python
# pipeline
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
classifier("I really enjoyed the movie, It was awesome")
```

```python
classifier("Total waste of time and money, do not watch this movie")
```

```python
review_list = ["I really enjoyed the movie, It was awesome",
               "Total waste of time and money, do not watch this movie"]

classifier(review_list)
```

```python
review_list = ["I really enjoyed the movie, It was awesome",
               "Total waste of time and money, do not watch this movie"]

classifier2(review_list)
```

## Dataset

```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("/content/drive/MyDrive/DL SK 21 3 23/NLP /Code and Data/review.csv")
df.head()
```

```python
movie_review_list = list(df["review"])
print(movie_review_list)
```

```python
result = classifier(movie_review_list)
result
```

```python
predictions = []

for op in result:
  if op["label"] == "NEGATIVE":
    predictions.append(0)
  else:
    predictions.append(1)

print(predictions)
```

```python
original_sentiment = df["sentiment"]

print(classification_report(original_sentiment, predictions))
```

## Zero-shot Classification

```python
zs_classifier = pipeline("zero-shot-classification")
```

```python
zs_classifier("This season IPL is very interesting, all matches are very close",
              candidate_labels=["cricket","football","basketball"])
```

```python
zs_classifier("Python is one of the best programming language",
              candidate_labels=["software","hardware","networking"])
```

```python
zs_classifier("You have won a lottery worth 3CR, click on the link to claim it",
              candidate_labels=["spam","ham"])
```
